```python
####  Esercizio numero 3 - Python  Clase1 #########
###   Guia Pagina 22 -----------------------

"""
Create un programma che permette di
inserire fino a 5 prodotti, con le seguenti
caratteristiche per ogni prodotto:
nome, prezzo, sconto e quantità acquistate.
Il programma ci restituirà in output il riepilogo,
totale da pagare e totale risparmiato.
"""
prodotti = {}
# Inserisci i dati da ogni articolo, fino a 5 totale:

num = 1
while num <= 5:
  print(f"Ciclo while: {num} ")
  print("A continuazione inserisci fino a 5 prodotti")
  prod_nome=input(f"inserisci il prodotto:")
  prod_prezzo=float(input(f"inserisci il prezzo del prodotto:"))
  prod_sconto=int(input(f"inserisci il sconto del prodotto in %:"))
  prod_quantita=int(input(f"inserisci la quantità del prodotto:"))
# Asegno a ogni variabili il numero de articolo inserito:
  prodotti[f"prodotti{num}"] = {
        "nome": prod_nome,
        "prezzo": prod_prezzo,
        "sconto": prod_sconto,
        "quantita": prod_quantita
  }
  num += 1

# variabili per totale:
totale_a_pagare = 0
totale_risparmiato = 0

# Calcoli totale:
for p in prodotti.values():
#    totale_parziale = 0
#    risparmio_parziale = 0
    subtotale = p["prezzo"] * p["quantita"]
    risparmio = subtotale * (p["sconto"] / 100)
#    risparmio_parziale = subtotale * (p["sconto"] / 100)
#    totale_parziale = (p["prezzo"] * p["quantita"]) - risparmio_parziale
    totale_a_pagare += subtotale - risparmio
    totale_risparmiato += risparmio

# Mostramos resumen
print("\n--- Resumen de la compra ---")
print("PRODOTTO     QUANTITA       PREZZO       SCONTO")
print("-----------------------------------------------")
for key, p in prodotti.items():

        print(f"{p['nome']}            {p['quantita']}          ${p['prezzo']}          {p['sconto']}%")
print(f"\nTotal a pagar: ${totale_a_pagare:.2f}")
print(f"Total ahorrado: ${totale_risparmiato:.2f}")


```
